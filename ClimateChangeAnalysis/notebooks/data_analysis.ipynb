```python
# Data Analysis and Visualization for Carbon Footprint

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv('../data/processed/cleaned_organization_emissions.csv')
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Emissions by sector
sector_emissions = df.groupby('Sector')['CO2_Emissions_tonnes'].sum().sort_values(ascending=False)
sector_emissions.plot(kind='bar', figsize=(8,4), title='Total CO2 Emissions by Sector')
plt.ylabel('CO2 Emissions (tonnes)')
plt.show()
```

```python
# Emissions by country
country_emissions = df.groupby('Country')['CO2_Emissions_tonnes'].sum().sort_values(ascending=False)
country_emissions.plot(kind='bar', figsize=(8,4), title='Total CO2 Emissions by Country')
plt.ylabel('CO2 Emissions (tonnes)')
plt.show()
```
